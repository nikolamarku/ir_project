```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df_test = pd.read_csv('testset.csv').drop('Unnamed: 0', axis=1)
df_test['type'] = 'test-set'
df_test.head()
```

```python
df_abstracts = pd.read_csv('abstracts.csv').drop('Unnamed: 0', axis=1)
df_abstracts['type'] = 'abstracts'
df_abstracts.head()
```

```python
import numpy as np
df = pd.concat([df_test,df_abstracts])
fig, ax = plt.subplots(figsize=(12, 4))
sns.barplot(
    data=df[df['sl']<=50],
    x="sl", y="wer", hue="type",hue_order=['test-set','abstracts'], errorbar=None
)
plt.ylabel('average wer')
plt.xlabel('sentence length (number of words)')
fig.savefig('average_wer.svg', format='svg')
```
